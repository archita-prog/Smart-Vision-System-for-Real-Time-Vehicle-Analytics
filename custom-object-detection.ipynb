```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")  # or yolov8m.pt if memory allows

model.train(
    data="/kaggle/input/car-detection1/data.yaml",
    epochs=100,
    imgsz=1024,
    batch=2,
    flipud=0.5,
    scale=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    name="yolov8_vehicle_1024",
    patience=0  # Early stopping if no improvement
)

```

```python
from ultralytics import YOLO

model = YOLO("runs/detect/yolov8_vehicle_1024/weights/best.pt")  # your trained weights

results = model.predict(
    source="/kaggle/input/car-detection1/test/images",  # or a single image path
    conf=0.25,  # confidence threshold
    save=True,  # saves output image(s) with bounding boxes
    name="yolov8_vehicle_test"
)
```
